```python
#Leitura das bibliotecas
import numpy as np
import matplotlib.pylab as plt

alfa = 1.0
beta = 2.4
N0 = np.zeros(20)
b1 = 10
N1 = np.zeros(20)
b2 = 15
N2 = np.zeros(20)
b3 = 20
N3 = np.zeros(20)
for var in range(20):
    N1[var] = (var/b1)**(1/(beta-alfa))
    N2[var] = (var/b2)**(1/(beta-alfa))
    N3[var] = (var/b3)**(1/(beta-alfa))
```

```python
plt.plot(N1,'*', label ='b= 10')
plt.plot(N2,'*', label ='b= 15')
plt.plot(N3,'*', label ='b= 20')
plt.plot(N0,'*')
plt.legend()
```

```python
#Leitura das bibliotecas
import numpy as np
import matplotlib.pylab as plt

alfa = 1.0
b = 10
N0 = np.zeros(20)
beta1 = 1.6
N1 = np.zeros(20)
beta2 = 2.5
N2 = np.zeros(20)
beta3 = 3.4
N3 = np.zeros(20)
for var in range(20):
    N1[var] = (var/b)**(1/(beta1-alfa))
    N2[var] = (var/b)**(1/(beta2-alfa))
    N3[var] = (var/b)**(1/(beta3-alfa))
```
