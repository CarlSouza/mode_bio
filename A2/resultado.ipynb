<h3><i> <font color="red">*Modelo Malthus </font></i></h3>

```julia
#constante de crescimento K>0
k=Sym("k")
#população inicial p(0)
p_0=Sym("p_0")
#P(t)=
p=p_0*exp(k*t)
```

```julia
dt=diff(p,t)
```

<p>Simulação durante um tempo t </p>

```julia
#Atribuindo valores
t= 0 : 0.5 : 40
k=4
p_0=100
p=p_0*exp.(k*t)
plot(p)
```

```julia
function G(t,k,p_0) # t deve ser colocado em forma de intervalo t=t_0:t_f
  p_0*exp.(k*t)
end
```

```julia
p=G(0:40,4,100)
plot(p)
```

```julia
dt_1=diff(p,t)
```
